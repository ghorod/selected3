```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv("LinksOfBooks.csv")
dataset = make_blobs(n_samples=6000,
                     centers=10,
                     n_features=2,
                     cluster_std=1.6,
                     random_state=50)
```

```python
points = dataset[0]
```

```python
kmeans = KMeans(n_clusters=10)
```

```python
kmeans.fit(points)
```

```python
plt.scatter(dataset[0][:,0],dataset[0][:,1])
```

```python
clusters = kmeans.cluster_centers_
```

```python
print(clusters)
```

```python
y_km = kmeans.fit_predict(points)
```
